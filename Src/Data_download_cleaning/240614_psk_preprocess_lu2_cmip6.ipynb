```r
crop_resam <- function(raster_data, ref) {
    r1 = crop(raster_data, ref)
    r2 = resample(r1, ref)
    return(r2)
}
```

```r
ssp2_guan_crop = crop_resam(ssp2_crop, ref_guan)
ssp2_guan_past = crop_resam(ssp2_past, ref_guan)
ssp2_guan_urbn = crop_resam(ssp2_urbn, ref_guan)
ssp2_guan_secd = crop_resam(ssp2_secd, ref_guan)
ssp2_guan_othr = crop_resam(ssp2_othr, ref_guan)

ssp5_guan_crop = crop_resam(ssp5_crop, ref_guan)
ssp5_guan_past = crop_resam(ssp5_past, ref_guan)
ssp5_guan_urbn = crop_resam(ssp5_urbn, ref_guan)
ssp5_guan_secd = crop_resam(ssp5_secd, ref_guan)
ssp5_guan_othr = crop_resam(ssp5_othr, ref_guan)

plot(ssp2_guan_past)
plot(ssp5_guan_secd)
```
