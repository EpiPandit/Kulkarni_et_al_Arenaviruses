```r
foi_func <- function(sdm, pop) {
  sdm <- raster(sdm)
  pop <- raster(pop)
  pop_crop <- crop(pop, sdm)
  pI = 0.0645
  pS = 0.95
  C = 1
  b = c(0.02/7, (0.0019 + 0.0029)/2, (1.76/(11*7) + 1.67/(11*7))/2)
  rat_den <- sdm
  values(rat_den) <- 0
  values(rat_den) <- rbinom(nrow(sdm)*ncol(sdm), 15, prob = values(sdm))
  foi_0.003 <- pop_crop
  values(foi_0.003) <- 0
  foi_0.003 <- (b[1]*pop_crop*rat_den)/(pop_crop + rat_den)
  names(foi_0.003) <- "b0.00286"
  foi_0.02 <- pop_crop
  values(foi_0.02) <- 0
  foi_0.02 <- (b[3]*pop_crop*rat_den)/(pop_crop + rat_den)
  names(foi_0.02) <- "b0.02227"
  foi <- brick(foi_0.003, foi_0.02)
  return(list(sdm = sdm, pop = pop_crop, rat_den = rat_den, foi = foi))
}
```

# Z.brevicauda

```r
foi_zyg <- foi_func(sdm = "./Data/Output/iterations/Z_brevicauda/current/probability_1.0_combined_nowater.tif",
         pop = "./Data/Input/Raw/World population/gpw_v4_2000.tif")
plot(brick(unlist(foi_zyg)))
```
