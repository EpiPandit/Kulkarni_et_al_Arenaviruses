```python
import os
from matplotlib import pyplot as plt
import rasterio
import shapely
import geopandas as gpd
import shutil
import glob
from sklearn.model_selection import train_test_split 
from pyimpute import load_training_vector
from pyimpute import load_targets
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
```

```python
os.chdir("/Users/pranavkulkarni/SDM/Climate_Models_Arenaviruses")
os.getcwd()
```

# zyg

```python
pa_files = sorted(glob.glob("./Data/Inter/paSampling/zyg_pa*.shp"))
pa_files
```

```python
len(pa_files)
```

```python
pa_list = []

for i in range(len(pa_files)):
    print(i)
    pa = gpd.GeoDataFrame.from_file(pa_files[i])
    pa_list.append(pa)
    print(pa_list[i].shape)
raster_features = sorted(glob.glob("./Data/Input/Processed/Resampled/guan/*.tif"))
print('number of raster features = ', len(raster_features))
```

```python
pa_train_list = []
pa_test_list = []

for i in range(50):
    pa_train, pa_test = train_test_split(pa_list[i], test_size=0.25, stratify=pa_list[i]['CLASS'])
    pa_train_list.append(pa_train)
    pa_test_list.append(pa_test)
    print(len(pa_train_list))
```

```python
%%time
train_xs_list = []
train_y_list = []
test_xs_list = []
test_y_list = []
test_target_xs_list = []
target_xs, raster_info = load_targets(raster_features)
target_xs[np.isnan(target_xs)] = 0

imputer = SimpleImputer(missing_values = np.nan, strategy = "mean")
for i in range(50):
    train_xs, train_y = load_training_vector(pa_train_list[i], raster_features, response_field = "CLASS")
    
    impute_train = imputer.fit(train_xs)
    train_xs = impute_train.transform(train_xs)
    train_xs_list.append(train_xs)
    train_y_list.append(train_y)
    test_xs, test_y = load_training_vector(pa_test_list[i], raster_features, response_field = "CLASS")
    impute_train = imputer.fit(test_xs)
    test_xs = impute_train.transform(test_xs)
    test_xs_list.append(test_xs)
    test_y_list.append(test_y)
    print(len(test_xs_list))
```

```python
raster_info
```

```python
%%time
np.savez_compressed("./Data/Inter/train_test_npz/zyg_train_xs", train_xs_list)
np.savez_compressed("./Data/Inter/train_test_npz/zyg_train_y", train_y_list)
np.savez_compressed("./Data/Inter/train_test_npz/zyg_target_xs", target_xs)
np.savez_compressed("./Data/Inter/train_test_npz/zyg_test_xs", test_xs_list)
np.savez_compressed("./Data/Inter/train_test_npz/zyg_test_y", test_y_list)
np.savez_compressed("./Data/Inter/train_test_npz/zyg_raster_info", [raster_info])
```
