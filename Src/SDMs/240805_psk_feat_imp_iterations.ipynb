```r
funct_sum <- function(data) {
  output <- data %>% 
    group_by(X) %>% 
    summarise(mean_fi = mean(X0, na.rm = T), 
              sd_fi = sd(X0, na.rm = T), 
              n_sel = n())
}
```

```r
RFE_zyg_rf <- read.csv("./Data/Output/iterations/Z_brevicauda/rf_results.csv")
noRFE_zyg_rf <- read.csv("./Data/Output/iterations/Z_brevicauda/rf_results_noRFE.csv")
RFE_zyg_rf_sum <- funct_sum(RFE_zyg_rf)
noRFE_zyg_rf_sum <- funct_sum(noRFE_zyg_rf)
zyg_sum <- full_join(RFE_zyg_rf_sum, noRFE_zyg_rf_sum, by = "X", suffix = c("_RFE", "_noRFE"))

write.csv(zyg_sum,"./Data/Output/iterations/Z_brevicauda/fi_summary.csv",row.names = F)
```
